```python
#The IQR is used to measure the spread of the middle 50% of data values. It is useful in identifying outliers and understanding the overall distribution of the dataset. Values that fall outside the range [Q1 - 1.5*IQR, Q3 + 1.5*IQR] are typically considered outliers.
#The first quartile (Q1) is the value below which 25% of the data falls, also known as the 25th percentile.
#The third quartile (Q3) is the value below which 75% of the data falls, also known as the 75th percentile.

Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3-Q1
print(IQR)
```

```python
#removing outliers

df = df[~ ((df<(Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]
df.shape
```

```python
#Histogram

df.Make.value_counts().nlargest(40).plot(kind='bar',figsize=(10,5))
plt.title("Number of cars by make")
plt.xlabel('Make')
plt.ylabel("No. of cars")
```

```python
#Heat Maps

plt.figure(figsize=(10,5))
c = df.corr()
sns.heatmap(c,cmap="BrBG", annot = True)
c
```
